```python
# why am I doing it like this lmao
x = 3
y = 1
```

```python
with open(filepath) as f:
    lines = f.readlines()
    
inputAsArray = []

for line in lines:
    inputAsArray.append(line)
```

```python
print(inputAsArray)
```

```python
treeCount = 0
startX = 0
startY = 0
while startY < len(inputAsArray):
    currentRow = inputAsArray[startY]
    currentPosition = currentRow[startX]
    if currentPosition == '#':
        treeCount+=1
    startX+=x
    width = len(inputAsArray[0])
    if startX >= width:
        startX-=width
    startY+=y
```
